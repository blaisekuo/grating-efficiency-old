```python
def PD_mean(str_run,PD):
    disp_list =[]

    for filestring in (glob.glob(datastore_path + str_run + "/*0.txt")):
        with open(filestring, 'r') as f:
            file2list = pd.read_csv(f,header=None,delimiter='\t')
        if PD == 1:
            for item in file2list[2]:
                disp_list.append(item)
        elif PD == 2:
            for item in file2list[4]:
                disp_list.append(item)
    
    mean = np.array(disp_list).mean()
    std = np.array(disp_list).std()
    
    print("mean: " + str(mean * 1000000) + ' \u03BCW')
    print("std: " + str(std * 1000000) + ' \u03BCW')
    
    return(np.array(disp_list).mean())
```

```python
plt.figure(figsize=(12, 12))    
      
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(True)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(True)    
     
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()    
  
plt.ylim(0, 100)    
#plt.xlim(2, 1278)        
  
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)    

plt.xlabel(r'$\gamma$ angle offset (degrees)',fontsize=20)
plt.ylabel("% efficiency",fontsize=20) 

plt.title(r'% efficiency vs. $\gamma$ angle',fontsize=25) 
  
plt.tick_params(axis="both", which="both", bottom="on", top="off",    
                labelbottom="on", left="on", right="off", labelleft="on")    


def plot_run(run_glob,lasermean,series_color,labelstring):
    
    i=0
    plot_list = []

    for file in (glob.glob(datastore_path + run_glob)):
        with open(file, 'r') as f:
            disp_list = pd.read_csv(f,header=None,delimiter='\t')

    #    purple_dot = plt.errorbar((np.arctan((10*i+20)/550))*(180/np.pi),(disp_list[2]/disp_list[4]).mean()*100,yerr=(disp_list[2]/disp_list[4]).std()*100,fmt='-o',color='purple')

        offplaneangle = (np.arctan((10*i+20)/550))*(90/np.pi)
        efficiency = (disp_list[2].mean()/lasermean)*100
        stddev = (disp_list[2].std()/lasermean)*100
        
        #print(file)
        #print(efficiency)
        #print(stddev)

        plot_list.append([offplaneangle,efficiency,stddev])

        i=i+1

    num_plot=np.array(plot_list)
    plotted = plt.errorbar(num_plot[:,0],num_plot[:,1],num_plot[:,2],fmt='-o',color=series_color,label = labelstring)
    
    return plotted

mean_run543 ="543nm_nosplit_lasercal_run14"
lasermean543=PD_mean(mean_run543,1)
mean_run633 ="633nm_nosplitter_lasercalpost_run21"
lasermean633=PD_mean(mean_run633,1)


green_dot = plot_run("543nm_nosplitter_run23/pos*_0.txt",lasermean543,"green",'543 nm, m=269')
purple_dot = plot_run("543nm_no1_nosplitter_run24/pos*_0.txt",lasermean543,"purple",'543 nm, m=268')

red_dot = plot_run("633nm_nosplitter_run19/pos*_0.txt",lasermean633,"red",'633 nm, m=231')
blue_dot = plot_run("633nm_no1_nosplitter_run20/pos*_0.txt",lasermean633,"blue",'633 nm, m=230')

plt.legend(loc='best')
#plt.legend([green_dot, purple_dot, red_dot, blue_dot],['543 nm, m=269', '543 nm, m=268', '633 nm, m=231', '633 nm, m=230'],loc='best')
```
