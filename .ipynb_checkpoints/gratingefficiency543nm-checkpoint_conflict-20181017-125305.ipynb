```python
plt.figure(figsize=(12, 12))    
      
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(True)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(True)    
     
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()    
  
plt.ylim(0, 100)    
#plt.xlim(2, 1278)        
  
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)    

plt.xlabel(r'$\gamma$ angle offset (degrees)',fontsize=20)
plt.ylabel("% efficiency",fontsize=20) 

plt.title(r'% efficiency vs. $\gamma$ angle',fontsize=25) 
  
plt.tick_params(axis="both", which="both", bottom="on", top="off",    
                labelbottom="on", left="on", right="off", labelleft="on")    


def plot_run(run_glob,series_color):
    
    i=0
    plot_list = []

    for file in (glob.glob(datastore_path + run_glob)):
        with open(file, 'r') as f:
            disp_list = pd.read_csv(f,header=None,delimiter='\t')

    #    purple_dot = plt.errorbar((np.arctan((10*i+20)/550))*(180/np.pi),(disp_list[2]/disp_list[4]).mean()*100,yerr=(disp_list[2]/disp_list[4]).std()*100,fmt='-o',color='purple')

        plot_list.append([(np.arctan((10*i+20)/550))*(90/np.pi),(disp_list[2]/disp_list[4]).mean()*100,(disp_list[2]/disp_list[4]).std()*100])

        i=i+1

    num_plot=np.array(plot_list)
    plotted = plt.errorbar(num_plot[:,0],num_plot[:,1],num_plot[:,2],fmt='-o',color=series_color)
    
    return plotted


green_dot = plot_run("543nm_run2/pos*_0.txt","green")
purple_dot = plot_run("543nm_on1_run2/pos*_0.txt","purple")
red_dot = plot_run("633nm_run2/pos*_1.txt","red")
blue_dot = plot_run("633nm_on1_run2/pos*_0.txt","blue")

plt.legend([green_dot, purple_dot, red_dot, blue_dot],['543 nm, m=0', '543 nm, m=0', '633 nm, m=1', '633 nm, m=1'],loc='upper right')
```
