# Grating Efficiency Measurements

## Setup

- Newport R6 echelle grating\
- 2mW 543 nm HeNe laser from Newport
- 0.8 mW 633 HeNe laser from Thorlabs
- both lasers attached to Fiber port collimators
- Corning 8.2 $\mu$m with 0.14 NA single mode fiber
- fiber attached to a Thorlabs 90$^o$ Off-axis Parabola that collimates a 4 mm beam
- the beam is shone on the channel 1 S120C photodiode (S/N 14120241) set to 10 1mW scale and sampling at 10 times per second
- power levels are measured using a Thorlabs Dual Channel Power and Energy Meter (PM320E) with two S120C photodiodes attached


## Method 1

Measure the power level of the two laser.
Check the run quality:
http://localhost:8888/lab/tree/gratingefficiency%2Frunquality.ipynb

Calcualte mean of power level.

Diffract beam from OAP with grating onto a position as close to the OAP as possible.
Measure the power level at this position and then at increasing off plane angles.

Plot grating efficiency as a percentage of the measured power level.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import re

#win
datastore_path="c:/cloudstor/datastore/grating-efficiency/"
```

```python
#read in csv file

with open(datastore_path + '633nm_run2/pos2_0.txt', 'r') as f:
    disp_list = pd.read_csv(f,header=None,delimiter='\t')

#disp_list=disp_list.drop(columns=['edge'])
    
#test
#print(disp_list[2].mean())
#print(disp_list[4].mean())

#print((disp_list[2]/disp_list[4]).mean())
#print(disp_list.std(axis=0))
#print(disp_list)
```

```python
def PD_mean(str_run,PD):
    disp_list =[]

    for filestring in (glob.glob(datastore_path + str_run + "/*0.txt")):
        with open(filestring, 'r') as f:
            file2list = pd.read_csv(f,header=None,delimiter='\t')
        if PD == 1:
            for item in file2list[2]:
                disp_list.append(item)
        elif PD == 2:
            for item in file2list[4]:
                disp_list.append(item)
    
    mean = np.array(disp_list).mean()
    std = np.array(disp_list).std()
    
    print("mean: " + str(mean * 1000000) + ' \u03BCW')
    print("std: " + str(std * 1000000) + ' \u03BCW')
    
    return(np.array(disp_list).mean())
```

# Laser Power Levels and Averages

## 543 nm laser

```python
mean_run543 ="543nm_nosplit_lasercal_run22"
lasermean543=PD_mean(mean_run543,1)
```

## 633 nm laser

```python
mean_run633 ="633nm_nosplitter_lasercalpost_run21"
lasermean633=PD_mean(mean_run633,1)
```

# Beam splitter ratio

## 543 nm

```python
run="543nm_lasercal_run29"
fullbeam=PD_mean(run,1)

run ="543nm_beamsplittercal_run30"

BS1_543=(PD_mean(run,1))/fullbeam
BS2_543=(PD_mean(run,2))/fullbeam

print("Beamsplitter grating position: " + str(BS1_543*100) + "%")
print("Beamsplitter reference position: " + str(BS2_543*100) + "%")
```

## 633 nm

```python
run="633nm_lasercal_run31"
fullbeam=PD_mean(run,1)

run ="633nm_beamsplittercal_run32"

BS1_633=(PD_mean(run,1))/fullbeam
BS2_633=(PD_mean(run,2))/fullbeam

print("Beamsplitter grating position: " + str(BS1_633*100) + "%")
print("Beamsplitter reference position: " + str(BS2_633*100) + "%")
```

```python
plt.figure(figsize=(12, 12))    
      
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(True)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(True)    
     
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()    
  
plt.ylim(0, 100)    
#plt.xlim(2, 1278)        
  
plt.yticks(np.arange(0,100,5),fontsize=15)
plt.xticks(fontsize=15)    

plt.xlabel(r'$\gamma$ angle offset (degrees)',fontsize=20)
plt.ylabel("% efficiency",fontsize=20) 

plt.title(r'% efficiency vs. $\gamma$ angle',fontsize=25) 
  
plt.tick_params(axis="both", which="both", bottom=True, top=False,    
                labelbottom=True, left=True, right=False, labelleft=True)    


def plot_run(run_glob,lasermean,series_marker,series_color,labelstring):
    
    i=0
    plot_list = []

    for file in (glob.glob(datastore_path + run_glob)):
        with open(file, 'r') as f:
            disp_list = pd.read_csv(f,header=None,delimiter='\t')

    #    purple_dot = plt.errorbar((np.arctan((10*i+20)/550))*(180/np.pi),(disp_list[2]/disp_list[4]).mean()*100,yerr=(disp_list[2]/disp_list[4]).std()*100,fmt='-o',color='purple')

        offplaneangle = (np.arctan((10*i+20)/550))*(90/np.pi)
        if lasermean == 0:
            if re.match('543',run_glob):
                efficiency = ( (disp_list[2]/disp_list[4]).mean() / (BS1_543/BS2_543) )*100
                stddev = ( (disp_list[2]/disp_list[4]).std() / (BS1_543/BS2_543) )*100
            elif re.match('633',run_glob):
                efficiency = ( (disp_list[2]/disp_list[4]).mean() / (BS1_633/BS2_633) )*100
                stddev = ( (disp_list[2]/disp_list[4]).std() / (BS1_633/BS2_633) )*100
            
        else:
            efficiency = (disp_list[2].mean()/lasermean)*100
            stddev = (disp_list[2].std()/lasermean)*100
        
        #print(file)
        #print(efficiency)
        #print(stddev)

        plot_list.append([offplaneangle,efficiency,stddev])

        i=i+1

    num_plot=np.array(plot_list)
    plotted = plt.errorbar(num_plot[:,0],num_plot[:,1],num_plot[:,2],fmt=series_marker,color=series_color,label = labelstring)
    
    return plotted


green_dot = plot_run("543nm_nosplitter_run23/pos*_0.txt",lasermean543,'-o',"green",'543 nm, m=269')

green_dash = plot_run("543nm_run8/pos*_0.txt",0,'--o',"green",'543 nm, m=269')
green_dash = plot_run("543nm_run2/pos*_0.txt",0,'--v',"green",'543 nm, m=269')

purple_dot = plot_run("543nm_no1_nosplitter_run24/pos*_0.txt",lasermean543,'-o',"purple",'543 nm, m=268')

purple_dash = plot_run("543nm_on1_run8/pos*_0.txt",0,'--o',"purple",'543 nm, m=268')
purple_dash = plot_run("543nm_on1_run2/pos*_0.txt",0,'--v',"purple",'543 nm, m=268')

red_dot = plot_run("633nm_nosplitter_run19/pos*_0.txt",lasermean633,'-o',"red",'633 nm, m=231')

red_dash = plot_run("633nm_run9/pos*_0.txt",0,'--o',"red",'633 nm, m=231')
red_dash = plot_run("633nm_run2/pos*_0.txt",0,'--v',"red",'633 nm, m=231')

blue_dot = plot_run("633nm_no1_nosplitter_run20/pos*_0.txt",lasermean633,'-o',"blue",'633 nm, m=230')

blue_dash = plot_run("633nm_on1_run9/pos*_0.txt",0,'--o',"blue",'633 nm, m=230')
blue_dash = plot_run("633nm_on1_run2/pos*_0.txt",0,'--v',"blue",'633 nm, m=230')

plt.legend(loc='best')
#plt.legend([green_dot, purple_dot, red_dot, blue_dot],['543 nm, m=269', '543 nm, m=268', '633 nm, m=231', '633 nm, m=230'],loc='best')
```
